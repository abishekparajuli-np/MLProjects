```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
X= car_df.drop(columns='Price',axis=1)
y = car_df['Price']
```

```python
car_df.columns
```

```python
oe=OneHotEncoder()
```

```python
car_df.dtypes
```

```python
from sklearn.compose import ColumnTransformer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
categorical_cols = X.select_dtypes(include=['object']).columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
pipe = Pipeline([
    ('preprocess', preprocessor),
    ('model', LinearRegression())
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score

cross_val_scores = cross_val_score(pipe, X_train, y_train, cv=5, scoring='r2')
print("Cross-Validation R Scores:", cross_val_scores)
print("Mean Cross-Validation R Score:", cross_val_scores.mean())
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
lr = LinearRegression(fit_intercept=True, positive=False)  # positive=False instead of normalize
ridge = Ridge(alpha=1.0, fit_intercept=True)
lasso = Lasso(alpha=0.1, max_iter=1000)
en = ElasticNet(alpha=0.1, l1_ratio=0.5, max_iter=1000)
dt = DecisionTreeRegressor(max_depth=None, min_samples_split=2)
rf = RandomForestRegressor(n_estimators=100, max_depth=None, random_state=42)
gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
ada = AdaBoostRegressor(n_estimators=50, learning_rate=1.0, random_state=42)
knn = KNeighborsRegressor(n_neighbors=5, weights="uniform")
svr = SVR(kernel="rbf", C=1.0, epsilon=0.1)
xgr=XGBRegressor()
```

```python
models = {
    "LR": lr,
    "Ridge": ridge,
    "Lasso": lasso,
    "EN": en,
    "DT": dt,
    "RF": rf,
    "GB": gb,
    "Ada": ada,
    "KNN": knn,
    "SVR": svr,
    "XGR":xgr
}
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.pipeline import Pipeline
import numpy as np

r2_scores = []
mae_scores = []
rmse_scores = []

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
for name, model in models.items():
    pipe = Pipeline([
        ('preprocess', preprocessor),  
        ('model', model)
    ])
        
    # Fit the pipeline
    pipe.fit(X_train, y_train)
        
    # Predict
    y_pred = pipe.predict(X_test)
        
    # Calculate metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
        
    # Cross-validated RMSE
    cv_scores = cross_val_score(pipe, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse = np.sqrt(-np.mean(cv_scores))
        
    print(f"Name: {name}\nR2: {r2:.4f}\nMAE: {mae:.4f}\nRMSE: {rmse:.4f}\n")
        
    # Append metrics
    r2_scores.append(r2)
    mae_scores.append(mae)
    rmse_scores.append(rmse)
```

```python
performance_df=pd.DataFrame({'Algorithm':models.keys(),'R2_Score':r2_scores,'MAE_Score':mae_scores,'RMSE_Score':rmse_scores})
```
