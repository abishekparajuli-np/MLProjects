```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("datasets/tmdb_5000_movies.csv")
credits=pd.read_csv("datasets/tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits,on="title")
```

```python
movies.head(1)
```

```python
#Important Columns (For Recommending Movies)
#Genre
#id
#keywords
#title
#overview
#cast
#crew
movies=movies[["movie_id","genres","keywords","title","overview","cast","crew"]]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]' 
#Now we need to convert this in the list of genre format
```

```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['genres'].head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['keywords'].head(1)
```

```python
movies.head(1)
```

```python
def top_three_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(top_three_cast)
```

```python
movies['cast'].head(1)
```

```python
def director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies.head(1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x ])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x ])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(3)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(3)
```

```python
movies_new_df=movies[['movie_id','title','tags']]
```

```python
movies_new_df.head()
```

```python
movies_new_df['tags']=movies_new_df['tags'].apply(lambda x:" ".join(x))
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_new_df['tags']=movies_new_df['tags'].apply(stem)
```

```python
movies_new_df['tags'][0]
```

```python
movies_new_df['tags']=movies_new_df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(movies_new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommendation(movie):
    movie_index= movies_new_df[movies_new_df['title'] == movie].index[0]
    distances= similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]    

    for i in movies_list:
        print(movies_new_df.iloc[i[0]].title)
  
```

```python
recommendation('Avatar')
```

```python
import pickle
```

```python
model_data = {
    "movies": movies_new_df,
    "similarity": similarity
}

with open("movie_recommendation.pkl", "wb") as f:
    pickle.dump(model_data, f)
```
